```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x=np.arange(0,10,1)
y=2*x+5
plt.plot(x,y,linestyle='--',linewidth=1.0,color='k',alpha=0.7,marker='D')
plt.title('Line plot demo')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend('Line1',loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x=np.arange(0,10,1)
y=2*x+5
z=3*x+5
plt.subplot(1,2,1)
plt.title('Graph 1')
plt.plot(x,y,linestyle='-.',marker='*',color='r')
plt.grid(True)
plt.subplot(1,2,2)
plt.title('Graph 2')
plt.plot(x,z,linestyle='--',marker='*',color='m')
plt.grid(True)
plt.show()
```

### boxplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
total=[1,5,2,7,3,9,10]
order=[54,23,6,7,9,41,5]
discount=[30,10,25,37,50,66,29]
data=list([total,order,discount])
plt.boxplot(data,showmeans=True)
plt.title('Boxplot demo')
plt.grid(True)
plt.show()
```

```python
import numpy as np
a=np.mean(total)
print(a)
b=np.mean(order)
print(b)
c=np.mean(discount)
print(c)
          
```

```python
import numpy as np
x=np.median(total)
print(x)
y=np.median(order)
print(y)
z=np.median(discount)
print(z)
```

### Pie chart

```python
animal=['Dog','Cat','Cow','Lion']
count=[40,20,50,28]
plt.pie(count,labels=animal,autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
animal=['Dog','Wolf','cat','lion']
count=[20,50,90,35]
plt.pie(count,labels=animal,autopct='%1.1f%%',shadow=True,startangle=90,explode=(0.2,0,0,0))
plt.title('Pie chart')
plt.show()
```

### Doughnut chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
group_names=['Group 1','Group 2','Group 3']
group_size=[20,30,50]
center=[10]
pie1=plt.pie(group_size,labels=group_names,radius=1.0,autopct='%1.1f%%',shadow=True,startangle=90)
pie2=plt.pie(center,radius=0.5,colors='w')
plt.title('Doughnut chart demo')
plt.show()
```

### Area chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=range(1,11)
y=[1,4,2,6,9,5,7,8,4,3]
plt.stackplot(x,y,color='red',alpha=0.3)
plt.plot(x,y,color='black',marker='o')
plt.title('Area chart demo')
plt.show()
```
